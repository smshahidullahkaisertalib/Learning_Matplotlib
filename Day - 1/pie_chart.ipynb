Pychart Plotting

```python
import numpy as np
import matplotlib.pyplot as plt

arr = np.array([3,4,3])
chart_label = ["Skyblue", "Orange", "Green"]
explodes = [0.01,0.01,0.01]

plt.pie(arr, labeldistance= 1.1, labels=chart_label, explode=explodes)
plt.title("Color Distribution")
# plt.legend()
plt.show()
```
