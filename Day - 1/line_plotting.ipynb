Line Plotting

```python
import matplotlib.pyplot as plt
import numpy as np

xp = np.array([0,2,4,6,8,10])
yp = np.array([0,3,4,8,12,16])

serif = {"family": "serif",
         "color": "blue",
         "size": 20
         }

plt.title("Python Plotting", fontdict= serif, loc="center")
plt.xlabel("horizontal")
plt.ylabel("vertical")

plt.grid()
plt.plot(xp, yp, "o", linestyle="solid", ms= 5, color="#08801c", mec = "#08801c", mfc = "#08801c")
plt.show()
```
