Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Pixel", "Xiaomi", "Samsung", "Apple", "Nokia"])
y = np.array([2,4,8,5,3])

plt.bar(x, y, color = "g", width= 0.5)
plt.title('Phone Users')
plt.xlabel("brand")
plt.ylabel("Users (in Millions)")
plt.grid(axis="y")
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np

# Defining the data
brands = np.array(["Pixel", "Xiaomi", "Samsung", "Apple", "Nokia"])
users = np.array([2, 4, 8, 5, 3])  # Convert '2M', '4M', etc., to integers

# Creating the bar chart
plt.barh(brands, users, color="g", height= 0.5)  # Bars are created for 'brands' on x-axis and 'users' on y-axis
plt.title('Phone Users')
plt.xlabel("Brand")
plt.ylabel("Users (in Millions)")
plt.show()
```
