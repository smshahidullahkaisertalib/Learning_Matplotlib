Sub Plotting


```python
import matplotlib.pyplot as plt
import numpy as np

xp = np.array([1,4,6,8,10])
yp = np.array([1,9,18.5, 15, 20])

plt.plot(xp, yp, marker="o", color= "g", ls="solid", ms = "5")
plt.title("Python Plotting")
plt.xlabel("Horizontal")
plt.ylabel("vertical")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# plot 1
xp = np.array([1,4,6,8,10])
yp = np.array([1,9,18.5, 15, 20])

plt.subplot(2,2,1)
plt.plot(xp, yp, marker="o", color= "g", ls="solid", ms = "5")
plt.title("Python Plotting 1")
plt.xlabel("Horizontal")
plt.ylabel("vertical")
plt.grid()

#plot 2

xp = np.array([1,4,6,8,10])
yp = np.array([1,9,18.5, 15, 20])

plt.subplot(2,2,2)
plt.plot(xp, yp, marker="o", color= "g", ls="solid", ms = "5")
plt.title("Python Plotting 2")
plt.xlabel("Horizontal")
plt.ylabel("vertical")
plt.grid()
plt.suptitle("Overall Report")
plt.show()
```
